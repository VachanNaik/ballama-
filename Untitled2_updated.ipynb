```python
import torch
import torch.nn as nn
```

```python
# Define vocabularies
amino_acids = "ACDEFGHIKLMNPQRSTVWY"
smiles_vocab = list("CNOPSHFIBrcl1234567890-=#()[]+/\\")
# Token dictionaries
aa_to_idx = {aa: i+1 for i, aa in enumerate(amino_acids)}
aa_to_idx["PAD"] = 0
smiles_to_idx = {ch: i+1 for i, ch in enumerate(smiles_vocab)}
smiles_to_idx["PAD"] = 0
```

```python
# class Attention(nn.Module):
#     def __init__(self, hidden_dim):
#         super().__init__()
#         self.attn = nn.Linear(hidden_dim, 1)

#     def forward(self, encoder_outputs):
#         # encoder_outputs: [batch_size, seq_len, hidden_dim]
#         weights = torch.softmax(self.attn(encoder_outputs).squeeze(-1), dim=1)  # [batch_size, seq_len]
#         context = torch.sum(encoder_outputs * weights.unsqueeze(-1), dim=1)     # [batch_size, hidden_dim]
#         return context

# for teturning the weights for printing the attention weights graphs
class Attention(nn.Module):
    def __init__(self, hidden_dim):
        super().__init__()
        self.attn = nn.Linear(hidden_dim, 1)

    def forward(self, encoder_outputs, return_weights=False):
        scores = self.attn(encoder_outputs).squeeze(-1)
        weights = torch.softmax(scores, dim=1)
        context = torch.sum(encoder_outputs * weights.unsqueeze(-1), dim=1)
        if return_weights:
            return context, weights
        return context
```

```python

# # Scratch DTI model
# class ScratchDTIModel(nn.Module):
#     def __init__(self, aa_vocab_size, smiles_vocab_size, emb_dim=64, hidden_dim=128):
#         super().__init__()
#         self.aa_embedding = nn.Embedding(aa_vocab_size, emb_dim, padding_idx=0)
#         self.smiles_embedding = nn.Embedding(smiles_vocab_size, emb_dim, padding_idx=0)

#         self.aa_encoder = nn.GRU(emb_dim, hidden_dim, batch_first=True)
#         self.smiles_encoder = nn.GRU(emb_dim, hidden_dim, batch_first=True)

#         self.mlp = nn.Sequential(
#             nn.Linear(2 * hidden_dim, 128),
#             nn.ReLU(),
#             nn.Linear(128, 1)
#         )

#     def forward(self, aa_seq, smiles_seq):
#         aa_emb = self.aa_embedding(aa_seq)
#         smiles_emb = self.smiles_embedding(smiles_seq)

#         _, aa_hidden = self.aa_encoder(aa_emb)
#         _, smiles_hidden = self.smiles_encoder(smiles_emb)

#         combined = torch.cat((aa_hidden[-1], smiles_hidden[-1]), dim=1)
#         return self.mlp(combined).squeeze()

# model = ScratchDTIModel(aa_vocab_size=len(aa_to_idx), smiles_vocab_size=len(smiles_to_idx))
# model.eval()


class ScratchDTIModelWithAttention(nn.Module):
    def __init__(self, aa_vocab_size, smiles_vocab_size, emb_dim=64, hidden_dim=128):
        super().__init__()
        self.aa_embedding = nn.Embedding(aa_vocab_size, emb_dim, padding_idx=0)
        self.smiles_embedding = nn.Embedding(smiles_vocab_size, emb_dim, padding_idx=0)

        # self.aa_encoder = nn.GRU(emb_dim, hidden_dim, batch_first=True, bidirectional=False)
        # self.smiles_encoder = nn.GRU(emb_dim, hidden_dim, batch_first=True, bidirectional=False)

        # adding bidrectional GRU
        self.aa_encoder = nn.GRU(emb_dim, hidden_dim, batch_first=True, bidirectional=True)
        self.smiles_encoder = nn.GRU(emb_dim, hidden_dim, batch_first=True, bidirectional=True)

        self.aa_attention = Attention(hidden_dim * 2)
        self.smiles_attention = Attention(hidden_dim * 2)

        self.mlp = nn.Sequential(
            nn.Linear(4 * hidden_dim, 128),  # 2 * (2 * hidden_dim)
            nn.ReLU(),
            nn.Linear(128, 1)
        )

        # self.aa_attention = Attention(hidden_dim)
        # self.smiles_attention = Attention(hidden_dim)

        # self.mlp = nn.Sequential(
        #     nn.Linear(2 * hidden_dim, 128),
        #     nn.ReLU(),
        #     nn.Linear(128, 1)
        # )



    # def forward(self, aa_seq, smiles_seq):
    #     aa_emb = self.aa_embedding(aa_seq)                    # [B, L, E]
    #     smiles_emb = self.smiles_embedding(smiles_seq)        # [B, L, E]

    #     aa_out, _ = self.aa_encoder(aa_emb)                   # [B, L, H]
    #     smiles_out, _ = self.smiles_encoder(smiles_emb)       # [B, L, H]

    #     aa_context = self.aa_attention(aa_out)                # [B, H]
    #     smiles_context = self.smiles_attention(smiles_out)    # [B, H]

    #     combined = torch.cat((aa_context, smiles_context), dim=1)
    #     return self.mlp(combined).squeeze()
    def forward(self, aa_seq, smiles_seq, return_attention=False):
          aa_emb = self.aa_embedding(aa_seq)
          smiles_emb = self.smiles_embedding(smiles_seq)

          aa_out, _ = self.aa_encoder(aa_emb)
          smiles_out, _ = self.smiles_encoder(smiles_emb)

          if return_attention:
            aa_context, aa_weights = self.aa_attention(aa_out, return_weights=True)
            smiles_context, smiles_weights = self.smiles_attention(smiles_out, return_weights=True)
          else:
            aa_context = self.aa_attention(aa_out)
            smiles_context = self.smiles_attention(smiles_out)

          combined = torch.cat((aa_context, smiles_context), dim=1)
          output = self.mlp(combined).squeeze()

          if return_attention:
            return output, aa_weights, smiles_weights
          return output
```

### ***Step 2: Normalize Binding Affinity (Y)***

```python
y_min = df["Y"].min()
y_max = df["Y"].max()
df["Y_scaled"] = (df["Y"] - y_min) / (y_max - y_min)
```

### ***Step 3: Create Dataset Class for PyTorch***

```python
from torch.utils.data import Dataset

class BindingAffinityDataset(Dataset):
    def __init__(self, dataframe, aa_vocab, smiles_vocab, max_aa_len=512, max_smiles_len=256):
        self.dataframe = dataframe
        self.aa_vocab = aa_vocab
        self.smiles_vocab = smiles_vocab
        self.max_aa_len = max_aa_len
        self.max_smiles_len = max_smiles_len

    def tokenize_sequence(self, seq, vocab, max_len):
        idxs = [vocab.get(ch, 0) for ch in seq[:max_len]]
        if len(idxs) < max_len:
            idxs += [0] * (max_len - len(idxs))
        return torch.tensor(idxs, dtype=torch.long)

    def __len__(self):
        return len(self.dataframe)

    def __getitem__(self, idx):
        row = self.dataframe.iloc[idx]
        aa_seq = self.tokenize_sequence(row["Target"], self.aa_vocab, self.max_aa_len)
        smiles_seq = self.tokenize_sequence(row["Drug"], self.smiles_vocab, self.max_smiles_len)
        # label = torch.tensor(row["Y"], dtype=torch.float)
        label = torch.tensor(row["Y_scaled"], dtype=torch.float)
        return aa_seq, smiles_seq, label
```

### 3. Move Model to **GPU** *italicized text*

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(device)
model = model.to(device)
```

### ***Define Optimizer and Loss***

```python
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)
criterion = torch.nn.MSELoss()
```

### ***5. Train the Model***

```python
epochs = 100
for epoch in range(epochs):
    model.train()
    total_loss = 0
    for aa_seq, smiles_seq, labels in train_loader:
        aa_seq, smiles_seq, labels = aa_seq.to(device), smiles_seq.to(device), labels.to(device)

        optimizer.zero_grad()
        preds = model(aa_seq, smiles_seq)
        loss = criterion(preds, labels)
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    print(f"Epoch {epoch+1}, Training Loss: {total_loss / len(train_loader):.4f}")
```

```python
def predict_all_pairs(smiles_list, protein_list, model, smiles_vocab, aa_vocab, y_min, y_max, device):
    model.eval()
    results = []

    with torch.no_grad():
        for smi in smiles_list:
            smiles_input = tokenize_sequence(smi, smiles_vocab, max_len=256).unsqueeze(0).to(device)

            for seq in protein_list:
                aa_input = tokenize_sequence(seq, aa_vocab, max_len=512).unsqueeze(0).to(device)

                pred_scaled = model(aa_input, smiles_input)
                pred_real = pred_scaled.item() * (y_max - y_min) + y_min

                results.append({
                    "smiles": smi,
                    "sequence": seq,
                    "scaled_prediction": pred_scaled.item(),
                    "real_prediction": pred_real
                })

    return results
```

```python
import ast
import numpy as np
import torch

def add_affinities_to_trials(df, model, smiles_vocab, aa_vocab, y_min, y_max, device):
    model.eval()
    avg_affinities = []
    max_affinities = []

    for _, row in df.iterrows():
        # Parse strings to lists if needed
        try:
            smiles_list = ast.literal_eval(row['smiless']) if isinstance(row['smiless'], str) else row['smiless']
            protein_list = ast.literal_eval(row['target_sequence']) if isinstance(row['target_sequence'], str) else row['target_sequence']
        except Exception as e:
            print(f"Error parsing SMILES or sequence for row {row['nctid']}: {e}")
            avg_affinities.append(None)
            max_affinities.append(None)
            continue

        affinities = []

        with torch.no_grad():
            for smi in smiles_list:
                smiles_input = tokenize_sequence(smi, smiles_vocab, max_len=256).unsqueeze(0).to(device)
                for seq in protein_list:
                    aa_input = tokenize_sequence(seq, aa_vocab, max_len=512).unsqueeze(0).to(device)
                    pred_scaled = model(aa_input, smiles_input)
                    pred_real = pred_scaled.item() * (y_max - y_min) + y_min
                    affinities.append(pred_real)

        if affinities:
            avg_affinities.append(np.mean(affinities))
            max_affinities.append(np.max(affinities))
        else:
            avg_affinities.append(None)
            max_affinities.append(None)

    df['avg_affinity'] = avg_affinities
    df['max_affinity'] = max_affinities
    return df
```

```python
import pandas as pd

# Load your CSV
df = pd.read_csv("sample_data/Book1.csv")

# Add affinity predictions
df_with_affinities = add_affinities_to_trials(df, model, smiles_to_idx, aa_to_idx, y_min, y_max, device)

# Save or inspect
df_with_affinities.to_csv("phase1_train_with_affinities.csv", index=False)
df_with_affinities.head()
```
